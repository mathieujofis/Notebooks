```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Good article on SGD: https://realpython.com/gradient-descent-algorithm-python/
```

```python
# Create a dataset and visualize it
X = pd.Series([np.random.uniform(0, 100) for i in range(0, 1000)])
y = pd.Series([2*x*np.random.uniform(1,1.1) + 3+np.random.normal(-15, 30)  for x in X])
train_df = pd.DataFrame(data = {'Input': X, 'Output': y})
sns.relplot(data=train_df, x='Input', y='Output')
```

```python
# Define a linear model with parameters to fit to the dataset
# f(x) = ax + b -- find parameters a and b (spoiler alert: it should be 7 and 3, respectively)
```

```python
# What is the loss function that needs to be minimized? MSE
# l = 1/N * sum[(y_actual - y_pred)^2]
#   = 1/N * sum[(y_actual - (ax+b))^2]

# Start by calculating the partial derivative for each parameter in the loss function
# Use the chain rule: d/dx f(g(x)) = f'(g(x))*g'(x)
# d/da y = 1/N *sum[-2x(y_actual - (ax + b))]
# d/db y = 1/N * sum[-2(y_actual - (ax + b))]
```

```python
# Implement gradient descent function to minimize loss function in batch
def calc_gradient(inputs, outputs, a, b):
    N = len(inputs)
    
    d_da = 0
    d_db = 0
    for i in range(N):
        x = inputs[i]
        y = outputs[i]

        d_da += -2*x*(y - (a*x + b))
        d_db += -2*(y - (a*x + b))
    
    # Divide to get average 
    grad_a = (1/float(N))*d_da
    grad_b = (1/float(N))*d_db
    
    return grad_a, grad_b

# Perform as many epochs as needed
def gradient_descent(inputs, outputs, alpha=.0002, a=0, b=0, num_epochs=1000):
    for i in range(num_epochs):
        grad_a, grad_b = calc_gradient(inputs, outputs, a, b)
        
        # After gradient vector is found, adjust scale (aka the learning rate). 
        # Subtract because we are minimizing (gradient should point to the most decreasing slope).
        a = a - grad_a * alpha
        b = b - grad_b * alpha
        
        if i%100 == 0:
            print(f'Current weights: {a}, {b}')
        
    return a, b

coef_a, coef_b = gradient_descent(X, y)
```

```python
# Plot model's regression line
x1 = 0
x2 = 100
y1 = coef_a*x1+coef_b
y2 = coef_a*x2+coef_b

sns.relplot(data=train_df, x='Input', y='Output')
plt.plot([x1, x2], [y1, y2], color='r')
```
