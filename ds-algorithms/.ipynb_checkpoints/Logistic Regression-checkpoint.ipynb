```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def rename_label(label):
    if label == 'setosa':
        return 1
    else:
        return 2

iris_df = sns.load_dataset('iris')
iris_df['species'] = iris_df['species'].apply(rename_label)
sns.pairplot(iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']], hue="species", diag_kind="kde")
```

```python
# Logistic model
# 1 / (1 + e^-z)
```

```python
# Loss function: cross entropy (combined into single equation) -> -avg[ yi*log(y_hati) + (1-yi)*log(1-y_hati) ]
# Derivative for gradient descent -> xi*(y_hat-y)
# See: https://towardsdatascience.com/logistic-regression-from-scratch-69db4f587e17
```

```python
class LogisticRegression:
    def sigmoid(self, z): 
        return 1 / (1 + np.e**(-z))
    
    # Just for calculating loss to report progress
    def loss_function(self, X, y, weights):
        z = np.dot(X, weights)
        
        # Cross-entropy function. 
        class_0 = y * np.log(self.sigmoid(z))
        class_1 = (1 - y) * np.log(1 - self.sigmoid(z))
        return -sum(class_1 + class_0) / X.shape[0]
        
    def train(self, X, y, epochs=25, lr=0.05):        
        loss = []
        weights = np.random.rand(X.shape[1])
        N = len(X)
        
        for i in range(epochs): 
            # Predict
            y_hat = self.sigmoid(np.dot(X, weights))
        
            # Update gradient and weights
            # X.T (a matrix) * (y_hat-y) (a vector) === sum of xi(y_hat-y) (where xi is a feature vector)
            weights = lr * np.dot(X.T, y_hat - y) / N
            
            loss.append(self.loss_function(X, y_hat, weights))
        
        self.weights = weights
        self.loss = loss
        print(loss)
            
    def predict(self, X):        
        # Predicting with sigmoid function
        z = dot(X, self.weights)
        # Returning binary result
        return [1 if i > 0.5 else 0 for i in self.sigmoid(z)]
        

lr = LogisticRegression()
y = iris_df['species']
X = iris_df.drop('species', axis=1)

lr.train(X, y)
```
